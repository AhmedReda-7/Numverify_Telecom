```sql
CREATE OR REPLACE TABLE etl.default.silver_customers AS
SELECT
    customer_id,
    CONCAT(first_name, ' ', last_name) AS customer_name,
    phone_number,
    country_code,
    country_name,
    line_type,
    carrier,
    plan_type,
    contract_type,
    payment_method,
    signup_date,
    tenure_months,
    CASE
    WHEN is_churned = 1 THEN 'True'
    WHEN is_churned = 0 THEN 'False'
    ELSE 'Not Defined'
    END AS is_churned
FROM etl.default.bronze_customers
WHERE customer_id IS NOT NULL;
```

```sql
select * from etl.default.silver_customers;
```

```sql
CREATE OR REPLACE TABLE etl.etl_schema.silver_biling AS
SELECT
    usage_id,
    customer_id,
    phone_number,
    billing_month,
    minutes_used,
    data_used_gb,
    sms_sent,
    monthly_charge_usd
FROM etl.etl_schema.bronze_biling
WHERE usage_id IS NOT NULL;
```

```sql
select * from etl.etl_schema.silver_biling;
```

```sql
CREATE OR REPLACE TABLE etl.etl_schema.silver_phone_data AS
SELECT
    carrier,
    country_code,
    country_name,
    country_prefix,
    international_format,
    line_type,
    local_format,
    valid
FROM etl.etl_schema.bronze_phone_data
Where carrier IS NOT NULL;
```

```sql
select * from etl.etl_schema.silver_phone_data;
```

```sql
CREATE OR REPLACE TABLE etl.etl_schema.silver_phone_dataa AS
SELECT
    CASE
    WHEN carrier = '' THEN 'confidential'
    ELSE carrier
    END AS carrier,
    country_code,
    country_name,
    country_prefix,
    international_format,
    line_type,
    local_format,
    valid
FROM etl.etl_schema.silver_phone_data_modify
where line_type is not null;
```

```sql
select * from etl.etl_schema.silver_phone_dataa;
```

```sql
drop table etl.etl_schema.silver_phone_data
```
